```python
import numpy as np
import matplotlib.pyplot as plt

# Define the colors of the cards
colors = ['red', 'blue', 'green', 'yellow']

# Define the probability distribution of each color
color_probabilities = [0.25, 0.35, 0.2, 0.2]

# Number of simulations
num_simulations = 1000

# Perform Monte Carlo simulation
results = np.random.choice(colors, size=num_simulations, p=color_probabilities)

# Count the occurrences of each color
color_counts = {color: np.sum(results == color) for color in colors}

# Calculate the probabilities
probabilities = {color: count / num_simulations for color, count in color_counts.items()}

# Plot the results
plt.bar(probabilities.keys(), probabilities.values(), color=probabilities.keys())
plt.xlabel('Colors')
plt.ylabel('Probability')
plt.title('Probability of Drawing Each Color')
plt.show()
```
