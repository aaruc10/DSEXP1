```python
import numpy as np
import matplotlib.pyplot as plt

# Number of coin flips in each simulation
num_flips = 1000

# Number of simulations
num_simulations = 1000

# Perform Monte Carlo simulation
results = np.random.choice(['H', 'T'], size=(num_simulations, num_flips))

# Count the occurrences of each outcome (H and T) across all simulations
head_counts = np.sum(results == 'H', axis=1)
tail_counts = np.sum(results == 'T', axis=1)

# Calculate the proportion of heads (H) and tails (T) for each simulation
head_probabilities = head_counts / num_flips
tail_probabilities = tail_counts / num_flips

# Plot the histogram of head and tail probabilities
plt.hist(head_probabilities, bins=30, alpha=0.5, color='b', label='Head Probabilities')
plt.hist(tail_probabilities, bins=30, alpha=0.5, color='r', label='Tail Probabilities')
plt.xlabel('Probability')
plt.ylabel('Frequency')
plt.title('Histogram of Head and Tail Probabilities')
plt.legend()
plt.show()
```
