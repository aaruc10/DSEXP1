```python
import numpy as np
import matplotlib.pyplot as plt

# Number of dice rolls in each simulation
num_rolls = 1000

# Number of simulations
num_simulations = 1000

# Perform Monte Carlo simulation
results = np.random.randint(1, 7, size=(num_simulations, num_rolls))

# Count the occurrences of each number (1 to 6) across all simulations
counts = [np.sum(results == i, axis=1) for i in range(1, 7)]

# Calculate the proportion of each number for each simulation
probabilities = [count / num_rolls for count in counts]

# Plot the histogram of probabilities for each number
plt.hist(probabilities, bins=30, alpha=0.5, label=['1', '2', '3', '4', '5', '6'])
plt.xlabel('Probability')
plt.ylabel('Frequency')
plt.title('Histogram of Dice Roll Probabilities')
plt.legend(title='Number')
plt.show()
```
